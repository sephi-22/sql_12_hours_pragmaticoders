```python
%load_ext sql
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .appName("example") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .config("spark.hadoop.validateOutputSpecs", "false") \
    .getOrCreate()
```

```python
df1 = spark.read.option("header", "true").option("inferSchema", "true").csv("NYC311ServiceRequestsDate02112023Start32436001toEnd33337000.csv")
df2 = spark.read.option("header", "true").option("inferSchema", "true").csv("NYC311ServiceRequestsDate02112023Start33337001toEnd34238000.csv")
df3 = spark.read.option("header", "true").option("inferSchema", "true").csv("NYC311ServiceRequestsDate02112023Start34238001toEnd35139000.csv")
```

```python
df = df1.union(df2).union(df3)
```

```python
%%sql_show
WITH BoroughComplaints AS    (SELECT 
        Borough, 
        `Complaint Type`, 
        COUNT(*) AS Complaint_Count, 
        ROW_NUMBER() OVER (PARTITION BY Borough ORDER BY COUNT(*) DESC) AS Rank
    FROM 
        myTable
    GROUP BY 
        Borough, `Complaint Type`)
SELECT * FROM BoroughComplaints WHERE Rank <=5
```

```python
%%sql_show
WITH YearlyComplaints AS (
    SELECT 
        YEAR(FROM_UNIXTIME(UNIX_TIMESTAMP(`created date`, 'MM/dd/yyyy hh:mm:ss a'))) AS year, 
        COUNT(*) AS complaint_count
    FROM 
        myTable
    WHERE 
        `complaint type` = 'Noise'
    GROUP BY 
        YEAR(FROM_UNIXTIME(UNIX_TIMESTAMP(`created date`, 'MM/dd/yyyy hh:mm:ss a')))
)
SELECT 
    year, 
    complaint_count
FROM 
    YearlyComplaints
ORDER BY 
    year
```

```python
%%sql_show
SELECT DISTINCT t1.City, t1.`Incident Zip`, t2.median_time_hours_city, t3.median_time_hours_zip, t4.access_issue_count
FROM myTable t1
LEFT JOIN
(SELECT City,
    PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY DATEDIFF(TO_DATE(`Closed Date`, 'MM/dd/yyyy hh:mm:ss a'), TO_DATE(`Created Date`, 'MM/dd/yyyy hh:mm:ss a'))) AS median_time_hours_city
    FROM
        myTable
    GROUP BY
        City) t2
ON t1.City = t2.City
LEFT JOIN
(SELECT
    `Incident Zip`,
    PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY DATEDIFF(TO_DATE(`Closed Date`, 'MM/dd/yyyy hh:mm:ss a'), TO_DATE(`Created Date`, 'MM/dd/yyyy hh:mm:ss a'))) AS median_time_hours_zip
    FROM
        myTable
    GROUP BY
    `Incident Zip`) t3
ON t1.`Incident Zip` = t3.`Incident Zip`
LEFT JOIN
(SELECT
    City,
    COUNT(*) AS access_issue_count
    FROM
        myTable
    WHERE
        `Resolution Description` LIKE '%was not able to gain access%'
    GROUP BY
        City) t4
ON t1.City = t4.City
```

```python
%%sql_show
SELECT DISTINCT t1.City, t1.`Incident Zip`, t2.median_time_hours_city, t3.median_time_hours_zip, t4.access_issue_count, t5.incident_count
FROM myTable t1
LEFT JOIN
(SELECT City,
    PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY DATEDIFF(TO_DATE(`Closed Date`, 'MM/dd/yyyy hh:mm:ss a'), TO_DATE(`Created Date`, 'MM/dd/yyyy hh:mm:ss a'))) AS median_time_hours_city
    FROM
        myTable
    GROUP BY
        City) t2
ON t1.City = t2.City
LEFT JOIN
(SELECT
    `Incident Zip`,
    PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY DATEDIFF(TO_DATE(`Closed Date`, 'MM/dd/yyyy hh:mm:ss a'), TO_DATE(`Created Date`, 'MM/dd/yyyy hh:mm:ss a'))) AS median_time_hours_zip
    FROM
        myTable
    GROUP BY
    `Incident Zip`) t3
ON t1.`Incident Zip` = t3.`Incident Zip`
LEFT JOIN
(SELECT
    City,
    COUNT(*) AS access_issue_count
    FROM
        myTable
    WHERE
        `Resolution Description` LIKE '%was not able to gain access%'
    GROUP BY
        City) t4
ON t1.City = t4.City
INNER JOIN
( SELECT
        City, `Incident Zip`,
        COUNT(*) AS incident_count,
        RANK() OVER (PARTITION BY City ORDER BY COUNT(*) DESC) AS rank
    FROM
        myTable
    GROUP BY
        City, `Incident Zip`) t5
ON t1.`Incident Zip` = t5.`Incident Zip` AND t1.City = t5.City AND t5.rank<=10
ORDER BY incident_count DESC, City
```
